```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("anime.csv")
```

```python
def  extract_data(text):
    data = ""
    flag = False
    for i in text:
        if i == "(":
            flag = True
        elif i == ")" and flag == True:
            flag = False
        elif flag == True:
            data+=i
    return data
```

```python
df['Total_episodes'] = df['Title'].apply(extract_data)
```

```python
df
```

```python
def gg(text):
    ff = text.split(" ")
    return int(ff[0])
    
    
df['Total_episodes'] = df['Total_episodes'].apply(gg)
```

```python
df
```

```python
df.loc[df['Total_episodes'] == df['Total_episodes'].max()]['Title']
```

```python
def Time_stamp(text):
    data = ""
    for i in range(len(text)):
        if text[i] == "-":
            for j in range(i-9,i+10):
                data += text[j]
    return data
            
```

```python
df["Time_stamp"] = df['Title'].apply(Time_stamp)
```

```python
df
```

```python
def edit(text):
    j = text[-19:]
    return j

df['Time_stamp'] = df['Time_stamp'].apply(edit)
```
